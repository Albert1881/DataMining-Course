```python
import numpy as np
import pandas as pd
import chardet
```

## Q 1

```python
CSV_FILE_PATH = './HW1data.csv'
with open(CSV_FILE_PATH,'rb') as data:
    result = chardet.detect(data.read(100000))
data_encoding=result['encoding']
```

```python
df = pd.read_csv(CSV_FILE_PATH,encoding=data_encoding)
df.head()
```

## Q 2
### a

```python
missing_values_count = df.isnull().sum()
row_num = df.shape[0]
missing_values_percentage = np.round(missing_values_count[df.columns]/row_num*100,2)
print("Percentage of missing data columns: ")
print(missing_values_percentage)

total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()
print("Percentage of total missing data: ", np.round(total_missing/total_cells*100,2), "%")
```

```python
info_data=None #Refer to the table for basic information
patient_key = 'PatientId'
info_subset = ['Gender','Age','Neighbourhood']

for column in info_subset:
    df_temp=df.groupby([patient_key,column]).size()
    df_temp=df_temp.reset_index(name='times')   #Count the number of occurrences of key and column combinations
    df_temp['times_max']=df_temp.groupby(['PatientId']).times.transform(max)    #Gets the value that occurs the most times
    df_temp=df_temp[df_temp.times==df_temp.times_max].drop_duplicates(['PatientId'])    #Gets the row corresponding to the most occurrences
    df_temp.drop(['times','times_max'],axis=1,inplace=True)
    if info_data is None:
        info_data = df_temp
    else:
        info_data = pd.merge(info_data,df_temp,on=patient_key,how='outer')
info_data.columns=info_data.columns+'_info' #Different from the original field
info_data
```

### b

```python
notNull_subset=['PatientId', 'ScheduledDay', 'AppointmentDay', 'SMS_received', 'No-show']
notNull_df = df[df[notNull_subset].notnull().T.all()]
notNull_df
```

### c

```python
fillNull_df = pd.merge(notNull_df,info_data,left_on=patient_key,right_on=patient_key+'_info',how='left')
for column in info_subset:
    fillNull_df[column].fillna(fillNull_df[column+'_info'],inplace=True)
    fillNull_df.drop([column+'_info'],axis=1,inplace=True)
fillNull_df.drop([patient_key+'_info'],axis=1,inplace=True)
fillNull_df
```

## Q 3

```python
notNull_subset=['PatientId', 'Gender', 'Age', 'Neighbourhood', 'ScheduledDay', 'AppointmentDay', 'SMS_received', 'No-show']
fillNull_df = fillNull_df[fillNull_df[notNull_subset].notnull().T.all()]
fillNull_df
```

## Q 4

```python
setZero_subset=['Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap']
for column in setZero_subset:
    fillNull_df[column].fillna(0,inplace=True)

fillNull_df
```

## Q 5

```python
useless_subset=['PatientId', 'AppointmentID']
for column in useless_subset:
    fillNull_df.drop([column],axis=1,inplace=True)
fillNull_df
```

## Q 6

```python
fillNull_df=fillNull_df[fillNull_df.Age>=0]
fillNull_df
```

## Q 7

```python
fillNull_df['AppointmentDay_prase'] = pd.to_datetime(fillNull_df['AppointmentDay'], infer_datetime_format = True, utc = True).dt.floor('D')
fillNull_df['ScheduledDay_prase'] = pd.to_datetime(fillNull_df['ScheduledDay'], infer_datetime_format = True, utc = True).dt.floor('D')
fillNull_df['Delta_Day']=(fillNull_df.AppointmentDay_prase-fillNull_df.ScheduledDay_prase).dt.days
fillNull_df=fillNull_df[fillNull_df['Delta_Day']>=0]

fillNull_df
```

## Q 8

```python
fillNull_df['SDay_DOW']=fillNull_df['ScheduledDay_prase'].dt.dayofweek
fillNull_df['ADay_DOW']=fillNull_df['AppointmentDay_prase'].dt.dayofweek
fillNull_df
```

## Q 9

```python
fillNull_df.drop(['AppointmentDay','AppointmentDay_prase','ScheduledDay','ScheduledDay_prase'],axis=1,inplace=True)
fillNull_df
```
