```python
for column in list(table_fillmean.columns[table_fillmean.isnull().sum() > 0]):
    if pd.api.types.is_numeric_dtype(table_fillmean[column]):
        mean_val = table_fillmean[column].mean()
        table_fillmean[column].fillna(mean_val, inplace=True)
```
