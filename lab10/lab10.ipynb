## DBSCAN
适合非凸数据聚类

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
from sklearn import neighbors
df = pd.read_csv('./DBSCAN_SAMPLE.csv')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
X = df.iloc[:,:].values
print(X)
```

```python
plt.scatter(X[:,0], X[:,1], marker=True)
plt.show()
```

```python
# standardize X
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=9)
y_pred = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.show()
# 可见 Kmeans 效果不好
```

```python
# 因此我们采用DBSCAN聚类算法
from sklearn.cluster import DBSCAN
dbscan = DBSCAN()
y_pred = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.show()
#默认参数聚类效果不佳
```

```python
# from sklearn.metrics import calinski_harabasz_score
# score = calinski_harabasz_score(X, y_pred)
# print("default parameter\n", score)
```

```python
dbscan1 = DBSCAN(eps=0.1)
y_pred = dbscan1.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score
score = calinski_harabasz_score(X, y_pred)
print("eps = 0.1\n", score)
```

```python
dbscan2 = DBSCAN(eps=0.1, min_samples=10)
y_pred = dbscan2.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score
score = calinski_harabasz_score(X, y_pred)
print("eps = 0.1, min_samples=10\n", score)
```

```python
for eps_ in [0.1, 0.2]:
    plt.subplot()
    dbscan_ = DBSCAN(eps=eps_)
    y_pred = dbscan_.fit_predict(X)
    plt.scatter(X[:,0], X[:,1], c=y_pred)
    score = calinski_harabasz_score(X, y_pred)
    plt.xlabel(" eps=" + str(eps_) + " score=" + str(score))
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
df_2 = pd.read_csv('./DBSCAN_TASK1.csv')
```

```python
# Let's create numpy arrays for features and target
X = df_2.iloc[:,:].values
```

```python
plt.scatter(X[:,0], X[:, 1], marker='o')
plt.show()
```

```python
# 因此我们采用DBSCAN聚类算法
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.08, min_samples=20)
y_pred = dbscan.fit_predict(X)
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.show()
```

```python
from sklearn.metrics import calinski_harabasz_score
score = calinski_harabasz_score(X, y_pred)
print("eps = 0.08, min_samples = 20\n", score)
```
