```python
import numpy as np
import pandas as pd
```

```python
TRAIN_FILE_PATH = './train.csv'
TEST_FILE_PATH = './test.csv'

train_dataset = pd.read_csv(TRAIN_FILE_PATH)
train_dataset.dropna(inplace=True)
```

```python
X_train = train_dataset.iloc[:, :-1].values.reshape(-1,1)
Y_train = train_dataset.iloc[:, 1:].values.reshape(-1,1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.scatter(X_train, Y_train, color='blue')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
LR_Model = LinearRegression()
```

```python
LR_Model.fit(X_train, Y_train)
print(LR_Model.intercept_)
print(LR_Model.coef_)
```

```python
w = LR_Model.coef_
b = LR_Model.intercept_
x = np.arange(X_train.min(), X_train.max(), 0.1)
y = w * x + b

plt.figure(figsize=(16,9))
plt.scatter(X_train, Y_train, color='blue')
plt.scatter(x, y, color='red')
plt.show()
```

## Evaluation

```python
from sklearn import metrics
```

```python
test_dataset = pd.read_csv(TEST_FILE_PATH)
test_dataset.dropna(inplace=True)
```

```python
X_test = test_dataset.iloc[:, :-1].values.reshape(-1,1)
Y_test = test_dataset.iloc[:, 1:].values.reshape(-1,1)
Y_pred = LR_Model.predict(X_test)
# scikit-learnMSE
print("MSE ",metrics.mean_squared_error(Y_test, Y_pred))
# scikit-learnRMSE
print("RMSE ",np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```
