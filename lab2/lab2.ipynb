```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup 
from time import sleep
```

```python
def scraping(url):
    html = urlopen(url) 
    soup = BeautifulSoup(html, 'html.parser') 
    for header in soup.find_all(class_='post-header'):
        info={}
        info['title']=header.h2.a.contents[0] #.text
        info['date']=header.find(class_='date').a.contents[0]
        info['author']=header.find(class_='author').a.contents[0]
        print(info)
    next_page = soup.find(class_='blog-pagination').find(class_='next-posts-link')

    if next_page:
        return next_page.get('href')
```

```python
url = "https://blog.scrapinghub.com" 
url_list = []
while url:
    sleep(1)
    if url in url_list:
        break
    else:
        url_list.extend(url)
        print(url)
    url=scraping(url)
    
```
